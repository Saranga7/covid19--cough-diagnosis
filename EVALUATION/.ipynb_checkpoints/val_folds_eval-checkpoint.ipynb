```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
path='validations'
```

```python
x=[]

for i, (dirpath, dirnames, filenames) in enumerate(os.walk(path)):
    print('Processing ', dirpath.split("/")[-1])
    for f in filenames:
        fpath=os.path.join(dirpath,f)
        count=0
        file=open(fpath,'r')
        while True:
            count += 1

            # Get next line from file
            line = file.readline()

            # if line is empty
            # end of file is reached
            if not line:
                break
            x.append(line.strip())
            print("{}".format(line.strip()))

        file.close()
```

```python
dicto={'file_name':x}
```

```python
df=pd.DataFrame(dicto)
```

```python
df.to_csv('Val_set.csv',index=False)
```
